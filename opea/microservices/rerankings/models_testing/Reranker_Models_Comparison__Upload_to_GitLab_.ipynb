# Metric

```python
from sklearn.metrics import ndcg_score
import numpy as np
```

```python
def get_ncdg_score(y_true, y_pred):
  return ndcg_score([y_true], [y_pred])
```

```python
def get_mrr_score(true_relevance, predicted_relevance):
    # Sort indices by predicted scores descending
    sorted_indices = np.argsort(predicted_relevance)[::-1]

    # Find the rank of the first truly relevant item
    for rank, idx in enumerate(sorted_indices, start=1):
        if true_relevance[idx] > 0:
            return 1 / rank
    return 0.0
```
